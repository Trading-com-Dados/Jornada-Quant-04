```python
#Primeiro criar a lista de preço de compra

lis_preco_compra = []

for i in range(13,len(btc_df)):
  if (btc_df['close'].iloc[i] > btc_df['MMA'].iloc[i]) and (btc_df['close'].iloc[i-1] < btc_df['MMA'].iloc[i-1]):
    preco_compra = btc_df['open'].iloc[i+1]
    lis_preco_compra.append(preco_compra)
  else:
    pass
```

```python
len(lis_preco_compra)
```

```python
lis_preco_venda = []

for i in range(13,len(btc_df)):
  if (btc_df['close'].iloc[i] < btc_df['MMA'].iloc[i]) and (btc_df['close'].iloc[i-1] > btc_df['MMA'].iloc[i-1]):
    preco_venda = btc_df['open'].iloc[i+1]
    lis_preco_venda.append(preco_venda)
  else:
    pass
```

```python
len(lis_preco_venda)
```

```python
lis_preco_venda.append('nao_fechou_ainda')
```

```python
df_lis_preco_compra = pd.DataFrame(lis_preco_compra, columns=['Preço_Compra'])
df_lis_preco_venda = pd.DataFrame(lis_preco_venda, columns=['Preco_Venda'])
df_lis_resultado = pd.concat([df_lis_preco_compra,df_lis_preco_venda], axis=1)
df_lis_resultado.head(50)
```

```python
df_lis_resultado['Preco_Venda'].iloc[36]
btc_df['close'].iloc[-1]
```

```python
df_lis_resultado['Preco_Venda'].iloc[36] = btc_df['close'].iloc[-1]
df_lis_resultado
```

```python
#Subtrair o preço de venda do preço de compra

df_lis_resultado['Resultado'] = df_lis_resultado['Preco_Venda'] - df_lis_resultado['Preço_Compra']
df_lis_resultado
```

```python
#Aqui calculamos a rentabilidade da estratégia em termos percentuais 

df_lis_resultado['Resultado_%'] = ((df_lis_resultado['Preco_Venda'] - df_lis_resultado['Preço_Compra'])/df_lis_resultado['Preço_Compra'])
df_lis_resultado
```

```python
#Calcular a evolução do patrimonio, vamos levar em consideração um caixa de 500000 e utilizar a função cumsum que acumula a soma dos resultados absolutos coluna 'Resultado'

caixa_inicial = 500000

df_lis_resultado['EvolucaoPatrimonio'] = df_lis_resultado['Resultado'].cumsum() + caixa_inicial
```

```python
df_lis_resultado
```

```python
#Faremos dois gráficos um de linha (Scatter) para a evoluçao do patrimônio e outro de barras para visualizar trades positivos e negativos

fig = make_subplots(rows=3, cols=1)
fig.add_trace(go.Scatter(x=df_lis_resultado.index, y=df_lis_resultado['EvolucaoPatrimonio'],name='Evolução Patrimonio'),row=1,col=1)
fig.add_trace(go.Bar(x=df_lis_resultado.index, y=df_lis_resultado['Resultado'], name='PnL'), row=2,col=1)
fig.update_yaxes(title_text="<b> Valor Caixa (R$)", row=1, col=1)
fig.update_yaxes(title_text="<b> PnL Ops (R$)", row=2, col=1)
fig.update_layout(title_text='Evolução do Patrimônio e PnL',width=1000 ,height=800)
fig.show()
```

```python
#Importando bibliotecas
import pandas as pd
import backtrader as bt
import time
import matplotlib.pyplot as plt
```

```python
#Criar novamente um dataframe com os mesmos parâmetros do backtesting realizado anteriormente

df = pd.DataFrame(client.get_historical_klines(symbol='BTCBRL',start_str='2021-01-01', end_str='2022-02-08', interval=client.KLINE_INTERVAL_1DAY))
df = df.iloc[:,:6]
df.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
df = df.set_index('date')
df.index = pd.to_datetime(df.index, unit='ms')
df = df.astype(float)
df
```

```python
#Datafeeds - onde inserimos os dados para realizar backtesting

data = bt.feeds.PandasData(dataname=df, name='BTCBRL')
```

```python
#Construir a estratégia dentro do backtrader de cruzamento de média e chamaremos de MMSCruzamento

class MMSCruzamento(bt.SignalStrategy):
  def __init__(self):
    mms = bt.ind.SMA(period=9)
    preco = self.data
    cruzamento = bt.ind.CrossOver(preco,mms)
    self.signal_add(bt.SIGNAL_LONG, cruzamento)
```

```python
#Executar a estratégia instanciando o cerebro e inserindo os parâmetros de caixa 500000, a estratégia MMSCruzamento e data

%matplotlib inline

cerebro = bt.Cerebro()
cerebro.broker.setcash(500000)
cerebro.addstrategy(MMSCruzamento)
cerebro.adddata(data)
cerebro.run()
plt.rcParams['figure.figsize'] = [10,7]
plt.rcParams.update({'font.size':12})
cerebro.plot(iplot=False)
```

```python
#Estrutura composta de duas funções, uma para alimentar e transformar os dados oriundos da API e as condições da estratégia e outa funçao que chama a anterior para executa-la através de iterações 

def order(coin, interval, quantity_trade):
  df = pd.DataFrame(client.get_klines(symbol=coin, interval=interval))
  df = df.iloc[:,:6]
  df.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
  df = df.set_index('date')
  df.index = pd.to_datetime(df.index, unit='ms')
  df = df.astype(float)
  df['mm_curta'] = df['close'].rolling(window=9).mean()
  time.sleep(5)
  print('fechamento anterior' + str(df.close[-2]) + ' ' + 'MediaMovel' + str(df.mm_curta[-2]))
  print('fechamento atual' + str(df.close[-1]) + ' ' + 'MediaMovel' + str(df.mm_curta[-1]))

  if (df.close[-1] > df.mm_curta[-1] and df.close[-2] < df.mm_curta[-2]):
    
    order = client.order_market_buy(
        symbol=coin,
        quantity=quantity_trade
    )

    wait_for_trade = False

    return order, wait_for_trade

  else:

    wait_for_trade = False
    order = None
    print('não executou' + str(df.index[-1]) + ' '+ 'Preço Atual'+str(df.close[-1]))

    return order, wait_for_trade

def send_order(coin, interval, quantity_trade, wait_for_trade, number_iteration):

  for i in range(number_iteration):
    wait_for_trade = True

    while wait_for_trade == True:

      trade, wait_for_trade = order(coin, interval, quantity_trade)
      print(trade)

```
